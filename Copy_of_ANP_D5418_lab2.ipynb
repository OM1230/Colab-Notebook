## Question 2
Take a number from the user and check whether it is:

- Positive
- Negative
- Zero

```python
number = int(input("Enter the number to check - "))
if number > 0:
  print("The number is positive")
elif number < 0:
  print("The number is negative")
else:
  print("The number is zero")
```

## Question 3
Take a number from the user and check:

- Is it divisible by 5?
- Is it divisible by 3?

Print True or False.

```python
number = int(input("Enter the number - "))
if number  % 5 == 0:
  print("The number is divisible by 5: True")
else:
  print("The number did not get divisible by 5: False")

if number % 3 == 0:
  print("The number is divisible by 3: True")
else:
  print("The number did not get divisible by 3: False")
```

## Question 4
Take a student's marks as input and print the grade:

Marks Range | Grade
------------|-------
90+ | A
75–89 | B
50–74 | C
Below 50 | Fail

```python
marks = int(input("Enter the marks: "))
if marks >= 90:
  print("The student got 'A' grade")
elif marks >= 75:
  print("The student got 'B' grade")
elif marks >= 50:
  print("The student got 'c' grade")
else:
  print("FAIL")
```

## Question 5
Take the age of a person and determine:

- Child (0–12)
- Teenager (13–19)
- Adult (20–59)
- Senior Citizen (60+)

```python
age = int(input("Enter the age: "))
if age <= 12:
  print("Hello Child")
elif age <= 19:
  print("Hello Teenager")
elif age <= 59:
  print("Hello Adult")
else:
  print("Hello Senior Citizen")
```
