### Question 4
Take two numbers from the user and print:

- Addition
- Subtraction
- Multiplication
- Division
- Modulus
- Exponent (power)

```python
val1 = int(input("Enter the number : "))
val2 = int(input("Enter the number : " ))
print(" ")
print("Addition = ", val1 + val2)
print("Subtraction = ", val1 - val2)
print("Multiplication = ", val1 * val2)
print("Division = ", val1 / val2)
print("Modulus = ", val1 % val2)
print("Exponent = ", val1 ** val2)
```
